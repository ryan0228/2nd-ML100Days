# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

筆記
標準差 https://zh.wikipedia.org/wiki/%E6%A8%99%E6%BA%96%E5%B7%AE


```python
pd.read_csv('./data/HomeCredit_columns_description.csv', encoding='latin1')

# 先篩選數值型的欄位
app_train.dtypes.value_counts()
```

```python
# 先篩選數值型的欄位
dtype_select = [np.dtype('int64'), np.dtype('float64')]

numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())>2 ))])
print("Numbers of remain columns %s" % len(numeric_columns))
```

```python

# 檢視這些欄位的數值範圍
# 底下用到的函數， matplotlib subplots boxplot 可以參考
# https://blog.csdn.net/tz_zs/article/details/81069499
# https://www.itread01.com/content/1541665023.html
# https://zhuanlan.zhihu.com/p/38199913

for col in numeric_columns:
    fig,ax = plt.subplots() #返回 多個子圖表 fig代表整個圖像，ax代表坐標軸和畫的圖
    ax.set_facecolor('#bbbbbb')#底圖顏色
    ax.boxplot(app_train[col].dropna(),vert=False) #vert : vert：是否需要將箱線圖垂直擺放，默認垂直擺放；
    ax.set_title(col) # 設定標題
    plt.show()
    plt.close()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# ECDF 經驗分布函數（英語：empirical distribution function）是統計學中一個與樣本經驗測度有關的分布函數
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
# 可以看出最大值是 【max      1.170000e+08】這數字太怪異

# 繪製 Empirical Cumulative Density Plot (ECDF)
amt_income_tot = app_train['AMT_INCOME_TOTAL']
cdf = amt_income_tot.value_counts().sort_index().cumsum()/len(amt_income_tot)
# cumsum （求累計次數）


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

## 補充：Normal dist 的 ECDF
![ecdf_normal](https://au.mathworks.com/help/examples/stats/win64/PlotEmpiricalCdfAndCompareWithSamplingDistributionExample_01.png)

```python
# 最大值落在分布之外
# REGION_POPULATION_RELATIVE 客戶居住區域的正常化人口數
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
# pandas 的value_counts()函數可以對Series裡面的每個值進行計數並且排序。
# pandas cumsum用法（求累計次數）
reg = app_train["REGION_POPULATION_RELATIVE"];
cdf = reg.value_counts().sort_index().cumsum()/len(reg)


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
# OBS_60_CNT_SOCIAL_CIRCLE 欄位是 客戶的回饋的意見數量

print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
obs = app_train['OBS_60_CNT_SOCIAL_CIRCLE']
cdf = obs.value_counts().sort_index().cumsum()/len(obs)


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```

## 注意：當 histogram 畫出上面這種圖 (只出現一條，但是 x 軸延伸很長導致右邊有一大片空白時，代表右邊有值但是數量稀少。這時可以考慮用 value_counts 去找到這些數值

```python
# 把一些極端值暫時去掉，在繪製一次 Histogram
# 選擇 OBS_60_CNT_SOCIAL_CIRCLE 小於 20 的資料點繪製
loc_a = app_train['OBS_60_CNT_SOCIAL_CIRCLE']<20
loc_b = 'OBS_60_CNT_SOCIAL_CIRCLE'
# loc     基於行標籤和列標籤（x_label、y_label）進行索引
app_train.loc[loc_a, loc_b].hist()
plt.show()
```
